```python
import os
import pydicom
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

def load_dicom_series(series_path):
    """Load a series of DICOM files from a folder and return as a 3D numpy array."""
    slices = [pydicom.dcmread(os.path.join(series_path, f)) for f in os.listdir(series_path) if f.endswith('.dcm')]

    # Sort slices based on their InstanceNumber
    slices.sort(key=lambda x: int(x.InstanceNumber))

    # Stack slices along the third axis to form a 3D volume
    volume = np.stack([s.pixel_array for s in slices], axis=-1)
    return volume

def preprocess_image(volume):
    """Normalize the MRI volume to the range [0, 1]."""
    volume = volume.astype(np.float32)
    volume = (volume - np.min(volume)) / (np.max(volume) - np.min(volume))
    return volume

def load_dataset(data_dir):
    """Load all series from the dataset directory."""
    images = []

    # Iterate through all folders inside the data directory
    for folder_name in os.listdir(data_dir):
        series_path = os.path.join(data_dir, folder_name)

        # Check if it's a directory and contains DICOM files
        if os.path.isdir(series_path):
            print(f"Processing folder: {folder_name}")

            try:
                # Load the DICOM series and preprocess it
                volume = load_dicom_series(series_path)
                volume = preprocess_image(volume)

                # Append the processed volume to the list
                images.append(volume)
            except Exception as e:
                print(f"Error processing folder {folder_name}: {e}")

    return np.array(images)

# Set your dataset directory path
data_dir = '/content/drive/MyDrive/RIDER_Breast_MRI_Dataset/Data'

# Load and preprocess the dataset
images = load_dataset(data_dir)

# Check the shape of the loaded dataset
print(f"Loaded {len(images)} MRI volumes")
print(f"Shape of first volume: {images[0].shape if len(images) > 0 else 'No data loaded'}")

# Display a sample slice from the first volume
if len(images) > 0:
    plt.imshow(images[0][:, :, images[0].shape[2] // 2], cmap='gray')
    plt.title("Sample MRI Slice")
    plt.show()


import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate, BatchNormalization, Activation, Add
from tensorflow.keras.models import Model

def conv_block(x, filters):
    """Residual Convolutional Block."""
    shortcut = x
    x = Conv2D(filters, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(filters, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    shortcut = Conv2D(filters, (1, 1), padding='same')(shortcut)
    x = Add()([x, shortcut])
    x = Activation('relu')(x)
    return x

def resunet(input_shape):
    """Builds the ResUNet model."""
    inputs = Input(input_shape)

    # Encoder
    c1 = conv_block(inputs, 32)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = conv_block(p1, 64)
    p2 = MaxPooling2D((2, 2))(c2)

    c3 = conv_block(p2, 128)
    p3 = MaxPooling2D((2, 2))(c3)

    c4 = conv_block(p3, 256)

    # Decoder
    u1 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c4)
    u1 = concatenate([u1, c3])
    c5 = conv_block(u1, 128)

    u2 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c5)
    u2 = concatenate([u2, c2])
    c6 = conv_block(u2, 64)

    u3 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c6)
    u3 = concatenate([u3, c1])
    c7 = conv_block(u3, 32)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c7)

    model = Model(inputs, outputs)
    return model

# Build the model
model = resunet((256, 256, 1))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()




def load_data(data_dir, mask_dir):
    """Load images and masks, preprocess them, and split into training/validation sets."""
    images = []
    masks = []

    # Check if the data directory exists
    if not os.path.exists(data_dir):
        print(f"Data directory does not exist: {data_dir}")
        return [], []

    for folder_name in os.listdir(data_dir):
        series_path = os.path.join(data_dir, folder_name)
        if os.path.isdir(series_path):
            print(f"Processing folder: {folder_name}")
            volume = load_dicom_series(series_path)

            # Skip if volume loading failed
            if volume is None:
                continue

            volume = preprocess_image(volume)

            # Adjust mask path: assuming masks are in a separate 'masks' directory
            mask_path = os.path.join(mask_dir, f"{folder_name}_mask.nii")  # Example for mask naming convention
            if os.path.exists(mask_path):
                mask = sitk.GetArrayFromImage(sitk.ReadImage(mask_path))
                images.append(volume)
                masks.append(mask)
            else:
                print(f"Mask not found for series: {folder_name}")

    # Check if any data was loaded
    if not images or not masks:
        print("No images or masks were loaded.")
        return [], []

    print(f"Loaded {len(images)} images and {len(masks)} masks")
    return np.array(images), np.array(masks)

# Usage example
data_dir = '/content/drive/MyDrive/RIDER_Breast_MRI_Dataset/Data'
mask_dir = '/content/drive/MyDrive/RIDER_Breast_MRI_Dataset/Masks'  # Adjust this path as needed
images, masks = load_data(data_dir, mask_dir)

# Split if data is successfully loaded
if len(images) > 0 and len(masks) > 0:
    from sklearn.model_selection import train_test_split
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)
    print(f"Training set: {len(X_train)}, Validation set: {len(X_val)}")
else:
    print("Failed to load dataset.")


#X_train, X_val, y_train, y_val = load_data(data_dir)



history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=4,
    validation_data=(X_val, y_val)
)

# Save the model
#model.save('/content/drive/MyDrive/resunet_tumor_segmentation.h5')


model.evaluate(X_val, y_val)
preds = model.predict(X_val)

plt.imshow(preds[0, :, :, 0], cmap='gray')
plt.title("Predicted Tumor Mask")
plt.show()

```

```python
import os
dataset_path = '/content/drive/MyDrive/kaggle_3m'
import cv2
import numpy as np
import os
from glob import glob
from sklearn.model_selection import train_test_split

# Define paths for images and masks
image_paths = glob(os.path.join(dataset_path, '*/*.tif'))
mask_paths = [p for p in image_paths if '_mask' in p]
image_paths = [p for p in image_paths if '_mask' not in p]

# Sort paths to align images and masks
image_paths.sort()
mask_paths.sort()
'''
# Function to load and preprocess data
def load_data(image_paths, mask_paths, image_size=(128, 128)):
    images, masks = [], []
    for img_path, mask_path in zip(image_paths, mask_paths):
        # Read image and mask
        img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)

        # Resize
        img = cv2.resize(img, image_size)
        mask = cv2.resize(mask, image_size)

        # Normalize image and mask
        img = img / 255.0
        mask = mask / 255.0
        mask = np.expand_dims(mask, axis=-1)  # Add channel dimension

        images.append(img)
        masks.append(mask)
    return np.array(images), np.array(masks)
'''

from concurrent.futures import ThreadPoolExecutor

def process_image_and_mask(image_path, mask_path, image_size):
    # Read image and mask
    img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)

    # Resize image and mask
    img = cv2.resize(img, image_size, interpolation=cv2.INTER_AREA)
    mask = cv2.resize(mask, image_size, interpolation=cv2.INTER_AREA)

    # Normalize image and mask
    img = img / 255.0
    mask = mask / 255.0
    mask = np.expand_dims(mask, axis=-1)  # Add channel dimension

    return img, mask

def load_data(image_paths, mask_paths, image_size=(128, 128)):
    # Use ThreadPoolExecutor for parallel processing
    with ThreadPoolExecutor() as executor:
        results = list(executor.map(
            process_image_and_mask, image_paths, mask_paths, [image_size] * len(image_paths)
        ))

    # Unpack results into separate lists
    images, masks = zip(*results)
    return np.array(images), np.array(masks)

# Load and preprocess data
image_size = (128, 128)  # Change size as needed
X, Y = load_data(image_paths, mask_paths, image_size=image_size)

# Split into training and validation sets
# First split: training and temporary (validation + test)
X_temp, X_test, Y_temp, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)  # 20% for testing

# Second split: training and validation
X_train, X_val, Y_train, Y_val = train_test_split(X_temp, Y_temp, test_size=0.25, random_state=42)  # 25% of temp for validation

print("Dataset split:")
print(f"Training set: {len(X_train)} images")
print(f"Validation set: {len(X_val)} images")
print(f"Test set: {len(X_test)} images")
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Activation, MaxPooling2D, Conv2DTranspose, Concatenate, Add
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import (
    Input, Conv2D, BatchNormalization, Activation, MaxPooling2D,
    Conv2DTranspose, Concatenate, Add, Dropout
)
def residual_block(x, filters, dropout_rate=0.2, l2_reg=0.01):
    shortcut = x
    x = Conv2D(filters, (3, 3), padding="same")(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = Dropout(dropout_rate)(x)
    x = Conv2D(filters, (3, 3), padding="same")(x)
    x = BatchNormalization()(x)
    shortcut = Conv2D(filters, (1, 1), padding="same")(shortcut)
    x = Add()([x, shortcut])
    x = Activation("relu")(x)
    return x

def ResUNet(input_shape):
    inputs = Input(input_shape)

    # Encoder
    e1 = residual_block(inputs, 64)
    p1 = MaxPooling2D((2, 2))(e1)

    e2 = residual_block(p1, 128)
    p2 = MaxPooling2D((2, 2))(e2)

    e3 = residual_block(p2, 256)
    p3 = MaxPooling2D((2, 2))(e3)

    # Bottleneck
    b = residual_block(p3, 512)

    # Decoder
    d1 = Conv2DTranspose(256, (2, 2), strides=(2, 2), padding="same")(b)
    d1 = Concatenate()([d1, e3])
    d1 = residual_block(d1, 256)

    d2 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding="same")(d1)
    d2 = Concatenate()([d2, e2])
    d2 = residual_block(d2, 128)

    d3 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding="same")(d2)
    d3 = Concatenate()([d3, e1])
    d3 = residual_block(d3, 64)

    outputs = Conv2D(1, (1, 1), activation="sigmoid")(d3)
    model = Model(inputs, outputs)
    return model

# Create model
input_shape = (image_size[0], image_size[1], 3)
model = ResUNet(input_shape)
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()

history = model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=25,
    batch_size=16
)

# Save model
model.save('/content/drive/MyDrive/ResUNet_ModelReg.h5')

# Evaluate
loss, accuracy = model.evaluate(X_val, Y_val)
print(f"Validation Loss: {loss}")
print(f"Validation Accuracy: {accuracy}")

# Predict on the test set
test_predictions = model.predict(X_test)
test_predictions = (test_predictions > 0.5).astype(np.float32)  # Convert to binary masks

# Evaluate metrics for the test set
test_results = []
for i in range(len(Y_test)):
    metrics = evaluate_metrics(Y_test[i], test_predictions[i])
    test_results.append(metrics)

# Compute average metrics across the test set
average_test_metrics = {k: np.mean([r[k] for r in test_results]) for k in test_results[0].keys()}
print("Average Metrics on Test Dataset:")
for metric, value in average_test_metrics.items():
    print(f"{metric}: {value:.4f}")

def dice_coefficient(y_true, y_pred):
    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()
    intersection = np.sum(y_true_flat * y_pred_flat)
    return (2.0 * intersection) / (np.sum(y_true_flat) + np.sum(y_pred_flat))

def iou_score(y_true, y_pred):
    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()
    intersection = np.sum(y_true_flat * y_pred_flat)
    union = np.sum(y_true_flat) + np.sum(y_pred_flat) - intersection
    return intersection / union

def precision(y_true, y_pred):
    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()
    true_positive = np.sum(y_true_flat * y_pred_flat)
    predicted_positive = np.sum(y_pred_flat)
    return true_positive / predicted_positive

def recall(y_true, y_pred):
    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()
    true_positive = np.sum(y_true_flat * y_pred_flat)
    actual_positive = np.sum(y_true_flat)
    return true_positive / actual_positive

def specificity(y_true, y_pred):
    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()
    true_negative = np.sum((1 - y_true_flat) * (1 - y_pred_flat))
    actual_negative = np.sum(1 - y_true_flat)
    return true_negative / actual_negative

# Evaluate metrics for the test set
def evaluate_metrics(y_true, y_pred):
    return {
        "Dice Coefficient": dice_coefficient(y_true, y_pred),
        "IoU Score": iou_score(y_true, y_pred),
        "Precision": precision(y_true, y_pred),
        "Recall": recall(y_true, y_pred),
        "Specificity": specificity(y_true, y_pred)
    }

# Calculate metrics for each test sample
test_results = []
for i in range(len(Y_test)):
    metrics = evaluate_metrics(Y_test[i], test_predictions[i])
    test_results.append(metrics)

# Compute average metrics across the test set
average_test_metrics = {k: np.mean([r[k] for r in test_results]) for k in test_results[0].keys()}
print("Average Metrics on Test Dataset:")
for metric, value in average_test_metrics.items():
    print(f"{metric}: {value:.4f}")
```
